### Data Visualization

### 1. Which team won most number of matches when they won the toss?

```python
ipl = pd.DataFrame(df)

# Filter the dataframe to get matches where the toss winner is the same as the match winner
winning_toss_and_match = ipl[ipl['toss_winner'] == ipl['winner']]

# Count the number of matches where toss winner also won the match by team
team_counts = winning_toss_and_match['winner'].value_counts()

# Create a bar plot
plt.figure(figsize=(10, 6))
team_counts.plot(kind='bar', color='skyblue')
plt.xlabel('Team')
plt.ylabel('Number of Matches')
plt.title('Number of Matches Where Toss Winner Also Won the Match by Team')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

### 2. Which batsman having highest runs in each year?

```python
sum_runs_each_year = ipl.groupby(['year', 'batsman'])['total_runs'].sum().reset_index()

# Get the index of the rows with the highest sum of runs for each year
idx = sum_runs_each_year.groupby('year')['total_runs'].idxmax()

# Filter the dataframe to get only the rows with the highest sum of runs for each year
result = sum_runs_each_year.loc[idx].dropna(axis=1, how='all')

print(result)
```

### 3. Which bowler having highest wickets in each year?

```python
sum_wickets_each_year = ipl.groupby(['year', 'bowler'])['is_wicket'].sum().reset_index()

# Get the index of the rows with the highest sum of wickets for each year
idx = sum_wickets_each_year.groupby('year')['is_wicket'].idxmax()

# Filter the dataframe to get only the rows with the highest sum of wickets for each year
result = sum_wickets_each_year.loc[idx].dropna(axis=1, how='all')

print(result)
```

### 4. Top 5 Stadiums with highest number of runs scored?

```python
venue_counts = ipl['venue'].value_counts().reset_index()
venue_counts.columns = ['venue', 'matches_played']

# Get the top 5 venues with the most matches played
top_5_venues = venue_counts.head(5)

print(top_5_venues)
```

### 5. From the above top 5 stadiums what is the final result if team chosse to bat first or field first?

```python
def plot_stadium_pie(stadium_name, ax):
    # Filter data for matches played at the specified stadium
    stadium_matches = df[df['venue'] == stadium_name]

    # Count the number of matches won when batting first and when fielding first
    bat_wins = stadium_matches[stadium_matches['toss_decision'] == 'bat'].shape[0]
    field_wins = stadium_matches[stadium_matches['toss_decision'] == 'field'].shape[0]

    # Calculate the total number of matches
    total_matches = len(stadium_matches)

    # Calculate the winning percentage when batting first and when fielding first
    bat_win_percentage = (bat_wins / total_matches) * 100
    field_win_percentage = (field_wins / total_matches) * 100

    # Create a pie chart to visualize the average winning percentage
    ax.pie([bat_win_percentage, field_win_percentage], labels=['Bat First', 'Field First'], autopct='%1.1f%%')
    ax.set_title(f'{stadium_name}')

# Create a figure with subplots for each stadium's pie chart
fig, axs = plt.subplots(1, 5, figsize=(20, 6))

# Call the function for each stadium to generate the pie charts side by side
plot_stadium_pie('Eden Gardens', axs[0])
plot_stadium_pie('Wankhede Stadium', axs[1])
plot_stadium_pie('Feroz Shah Kotla', axs[2])
plot_stadium_pie('Rajiv Gandhi International Stadium, Uppal', axs[3])
plot_stadium_pie('M Chinnaswamy Stadium', axs[4])

plt.tight_layout()
plt.show()
```
